```python
# %load hwk1_mbali.py
#!/usr/bin/env python3
"""
Created on Mon Mar  9 12:29:00 2020

@author: manik.bali
"""
#This script consists of answers to all the questions of HW1.
#It can be run on spyder or any python environment tool
#Question 1
#Problem 1. Assume you are given a list of points as follows:(2,2,2),(3,3,3),(4,4,4),(6,0,0),(0,6,0),(0,0,6).
#Write two functions, f indMean and f indStd, in Python that return the mean and the standard
#deviation of these points, respectively. Generalize your functions to accept a list of different
#number of points.

#Answer
#The key is to construct a rearrange the tuples into three tuples of lengh=number of tuples
# In this way all the numbers needed to compute mean and std will be in a single tuple
#If we do this we can pick one tuple and do math on it ( such as mean standard deviation) within it.
#zip(*lpoints) unzips and rearranges n tuples into three (dimension of a single tuple) tuples. To compute the sum of the elements of the tuple
#one can use the sum function and to compute mean and std. Also this method is scalable to any number of 
#points
from math import sqrt

lpoints=[(2,2,2),(3,3,3),(4,4,4),(6,0,0),(0,6,0),(0,0,6)]

x=sum(list(zip(*lpoints))[0])/len(list(zip(*lpoints))[0])
y=sum(list(zip(*lpoints))[1])/len(list(zip(*lpoints))[1])
z=sum(list(zip(*lpoints))[2])/len(list(zip(*lpoints))[2])

mean_point=[(x,y,z)]

var_x=list(map(lambda i: list(zip(*lpoints))[0][i] -x,range(len(list(zip(*lpoints))[0]))))
var_y=list(map(lambda i: list(zip(*lpoints))[0][i] -y,range(len(list(zip(*lpoints))[0]))))
var_z=list(map(lambda i: list(zip(*lpoints))[0][i] -z,range(len(list(zip(*lpoints))[0]))))
     
prod_X=list(map(lambda x,y: x*y,var_x,var_x) )
prod_Y=list(map(lambda x,y: x*y,var_y,var_y) )
prod_Z=list(map(lambda x,y: x*y,var_z,var_z) )

std_x=sqrt(sum(prod_X)/len(prod_X))
std_y=sqrt(sum(prod_Y)/len(prod_Y))
std_z=sqrt(sum(prod_Z)/len(prod_Z))

std_estimate=[(std_x,std_y,std_z)]

print("Question 1")
print("Mean Tuple=",mean_point)
print("Standard Deviation=",std_estimate)

# the function is general

#Question 2
#Write a Python function, InnerP roduct, to accept two lists (representing vectors) as
#arguments and return their inner (dot) product.

print("Question 2")

def innerproduct(a,b):
#Use map/lambda combination to construct inner product function and pass input lists     
    prod=list(map(lambda x,y: x*y,a,b) )
    return sum(prod)

a=[0,1,2,3,4]
b=[0,1,2,3,4]
c=innerproduct(a,b)
print("Inner Product=",c)

#Question 3 
#(a) The most popular boy name for each year between 1880 and 2008.
#(b) The most popular girl name for each year between 1880 and 2008.
#(c) The most popular name across all these years for both boys and girls based on the percentage of the population that has had that name.

#Question 3 -A B C. Print names of most common and least common baby names year wise
#The following part of the code gives output for Question 3 A B and C.
#This is controlled by setting Question=xx where xx takes values a, b and c3(a)
#



import sys
file = open(r"C:\Users\Navya\Downloads\baby-names-data.txt","r")
next(file) #skip header
lines = file.read().splitlines() # read all the lines of data into lines

#seperate the data into heads
year=list(map(lambda i: int(lines[i].split(',')[0]), range(len(lines))))
names=list(map(lambda i: lines[i].split(',')[1], range(len(lines))))
percent=list(map(lambda i: float(lines[i].split(',')[2]), range(len(lines))))
gender=list(map(lambda i: lines[i].split(',')[3], range(len(lines))))

Question="c"
if (Question=="c"):
        print("Question 3(c)")
        boy_idx=list(filter(lambda i: gender[i]=='"boy"', range(len(gender))))
        girl_idx=list(filter(lambda i: gender[i]=='"girl"', range(len(gender))))
        plist=list(map(lambda x: percent[x],boy_idx ))  
        boy_names= list(map(lambda x: names[x],boy_idx )) 
        max_idx=plist.index(max(plist))
        print("Most Popular Boy Name( 3C)=",boy_names[max_idx])
        plist=list(map(lambda x: percent[x],girl_idx ))
        girl_names= list(map(lambda x: names[x],girl_idx )) 
        max_idx=plist.index(max(plist))
        print("Most Popular Girl Name( 3C)=",girl_names[max_idx])
        sys.exit("Exiting Normally")
years=(list(set(year)))  # Get Uniq years
#year_idx=list(filter(lambda i: year[i]==, range(len(year))))


#Question 3 -A B C. Print names of most common and least common baby names year wise

#loop over years and identify max percentage of names
# three subcases ( each pertaining to the questions a,b,c)
for num, years in enumerate(years):   
#    year_idx=list(filter(lambda i: year[i]==years, range(len(year))))
    if ( Question=="a"):
        year_idx= list(filter(lambda i: ((year[i]==years) & ( gender[i]=='"boy"')), range(len(year))))
        if ( num==1):
            print("Question 3(a)",num)
    if (Question=="b"):    
        year_idx= list(filter(lambda i: ((year[i]==years) & ( gender[i]=='"girl"')), range(len(year))))
        if ( num==1):
            print("Question 3(b)",num)
         
                  
       
    plist=list(map(lambda x: percent[x],year_idx )) 
    nlist=list(map(lambda x: names[x],year_idx ))
    glist=list(map(lambda x: gender[x],year_idx ))
    
    max_percent=max(plist)
    min_percent=min(plist)
 
    max_idx=plist.index(max(plist))
    min_idx=plist.index(min(plist))

#    year_idx
    print(year[year_idx[10]],nlist[max_idx],glist[max_idx],max_percent)
#     year_idx=list(filter(lambda i: (year[i]==years) & ( gender[i]=="boy"), range(len(year))))
#def most_pop:
    

#print(names[i])
```
